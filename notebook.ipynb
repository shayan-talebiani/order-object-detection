```python
from ultralytics import YOLO

IMAGE_PATH = "images/1"

# Load your trained model
model = YOLO("trained-model/best.pt")

# Run prediction on an image or directory
results = model.predict(source=IMAGE_PATH, save=True)

process_results = []

# Loop through results
for r in results:
    process_r = []
    print("\n--- Detection Result ---")
    for box in r.boxes:
        #print(f"cls shape: {box.xyxy}")
        cls_id = int(box.cls[0])                # class id
        conf = float(box.conf[0])               # confidence
        xyxy = box.xyxy[0].tolist()             # [x1, y1, x2, y2]
        process_r.append({"id":cls_id , "xyxy":xyxy})
        print(f"Class ID: {cls_id}, Confidence: {conf:.2f}, Box: {xyxy}")
    process_results.append(process_r)

# Load your trained model
#model = YOLO("trained-model/best.pt")
"""
# Run prediction on an image or directory
results = model.predict(source="HTP_Dataset_yolov11_augmentation/train/images", save=True)

# Loop through results
for r in results:
    print("\n--- Detection Result ---")
    for box in r.boxes:
        cls_id = int(box.cls[0])                # class id
        conf = float(box.conf[0])               # confidence
        xyxy = box.xyxy[0].tolist()             # [x1, y1, x2, y2]

        print(f"Class ID: {cls_id}, Confidence: {conf:.2f}, Box: {xyxy}")

"""
```
